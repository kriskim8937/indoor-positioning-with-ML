```python
import pandas as pd
```

```python
#Read data
wifi_data = pd.read_csv("wifi_data.csv")
#wifi_data.head(5)
#wifi_data.isnull().sum()
```

```python
#Remove location_unknown
for index, row in wifi_data.iterrows():
    if(row['location_coded'] == "location_unknown") :
        wifi_data.drop(index, inplace=True)
#wifi_data.isnull().sum()
```

```python
#Fill all the Nan with -174
wifi_data_na_filled = wifi_data.fillna(-174)
print(len(wifi_data_na_filled))
#wifi_data_na_filled.isnull().sum()
```

```python
## Data Split
from sklearn.model_selection import train_test_split

X = wifi_data_na_filled.iloc[:,:-3]

Xplot = wifi_data_na_filled
y = wifi_data_na_filled['location_coded']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1235)
X_train_plot, X_test_plot, y_train_plot, y_test_plot = train_test_split(Xplot,y,test_size=0.3,random_state=1235)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_clf = RandomForestClassifier(n_estimators=100, random_state=1)
random_forest_clf.fit(X_train, y_train) 
print("""------------------- Random forest -------------------""")
print(classification_report(y_test,random_forest_clf.predict(X_test)))
```

```python
# good enough score for some locations 
# visulize the data by floating the coordinate according to the location, group some locations 
# gather some locations because there not may datas but they are quite similar
train_score = random_forest_clf.score(X_train,y_train)
test_score = random_forest_clf.score(X_test,y_test)
print("Accuracy train/test: %.2f/%.2f" % (train_score, test_score))
```
